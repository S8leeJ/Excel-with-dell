```python
import pandas as pd
from deep_translator import GoogleTranslator
from langdetect import detect

# 1. Load the Dataset
df = pd.read_csv("fusiontech_reviews.csv") # Ensure this file is in the same directory as your script

# Initialize the translator (Auto-detect source language, translate to English)
translator = GoogleTranslator(source='auto', target='en')

# 2. Define a function to safely translate text
def translate_to_english(text):
    if not isinstance(text, str) or len(text.strip()) < 3:
        return text # Return as-is if it's missing or too short
    
    try:
        # Detect the language first to save time/API calls
        lang = detect(text)
        if lang != 'en':
            # If it's not English (e.g., Spanish "es"), translate it
            return translator.translate(text)
        else:
            # If it's already English, leave it alone
            return text
    except Exception as e:
        # If langdetect or translation fails (e.g., weird characters), return original
        return text

# 3. Apply the translation to the Review Text and the Review Title
print("Translating review text... (This may take a minute or two depending on dataset size)")
df['text'] = df['text'].apply(translate_to_english)

print("Translating review titles...")
df['title_x'] = df['title_x'].apply(translate_to_english)

print("Translation Complete! Your dataset is now fully in English.")

# Optional: Save the translated dataset so you don't have to re-translate it every time
df.to_csv("FusionTech_Translated_Dataset.csv", index=False)
```
